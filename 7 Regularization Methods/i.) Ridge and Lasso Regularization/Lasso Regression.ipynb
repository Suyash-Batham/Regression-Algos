# L1 Regularization or Lasso Regression

###### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

###### Importing dataset

```python
dataset = pd.read_csv('Ridge.csv')
dataset
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

###### Importing Lasso Regression Algo

```python
from sklearn.linear_model import Lasso
```

```python
lr = Lasso(alpha=0.1)   # (alpha = 0.01,1,5)
lr.fit(X,y)
```

###### Visualizing Ridge Regression

```python
plt.scatter(X,y,color='red')
plt.plot(X,lr.predict(X),color='blue')
plt.title("Lasso Regression")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
x_plt = [0,1,2,3,4]
lasso_l = [0,0.5,1,2,4]
```

```python
for i,l in enumerate(lasso_l):
    lasso = Lasso(alpha=l)
    lasso.fit(X,y)
    
    lasso_coeff = lasso.coef_
    lasso_intercept = lasso.intercept_
    y_plt = lasso.predict(pd.DataFrame(x_plt))
    
    plt.figure(2)
    plt.plot(x_plt,y_plt)
    plt.xlim(xmin = 0, xmax= 6)
    plt.ylim(ymin = 0, ymax= 9)
    plt.text(x_plt[-1],y_plt[-1],
        ' y = '  +
        str('%.2f' %lasso_coeff) +
        ' * x + '     +
        str('%.2f' %lasso_intercept) +
        ' for \u03BB or \u03B1 = '  +
        str(l), 
        fontsize = 12 )
```
